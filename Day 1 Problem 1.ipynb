# Advent of Code - 2025 - Day 1 - Problem 1

https://adventofcode.com/2025/day/1

## Load Source Data

Load source data into `DATA`.

```python
f = open("data/day1.txt", "r")
DATA = list(map(str.strip, f.readlines()))
f.close()

DATA
```

## Process Instructions

Use the instructions to determine the password.

```python
current_position = 50
password = 0
for turn in DATA:
    direction = turn[:1]
    distance = int(turn[1:])
    if direction == 'L':
        distance = -distance
    current_position = (current_position + distance) % 100
    if current_position == 0:
        password += 1

password
```
