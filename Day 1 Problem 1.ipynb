## Load Source Data

Load source data into `DATA`.

```python
# Read the input file containing turn instructions
with open("data/day1.txt") as f:
    DATA = [line.strip() for line in f]

# DATA
```

## Process Instructions

Use the instructions to determine the password.

```python
# Start at position 50 on a circular dial (0-99)
current_position = 50
password = 0

for turn in DATA:
    # Parse the turn instruction: first character is direction (L/R), rest is distance
    direction = turn[:1]
    distance = int(turn[1:])
    
    # Convert left turns to negative distance for easier calculation
    if direction == 'L':
        distance = -distance
    
    # Move on the circular dial (wraps around at 100)
    current_position = (current_position + distance) % 100
    
    # Increment password counter each time we land exactly on zero
    if current_position == 0:
        password += 1

password
```
