## Create Graph and Find Minimum Distance

```python
def count_paths_dag(graph: dict[str, tuple[str, ...]], start: str, finish: str) -> int:
    """
    Counts paths in a DAG from start to finish using memoization.
    """
    memo: dict[str, int] = {}

    def dfs_memo(node: str) -> int:
        if node == finish:
            return 1
        if node in memo:
            return memo[node]

        count = 0
        if node in graph:
            for neighbor in graph[node]:
                count += dfs_memo(neighbor)

        memo[node] = count
        return count

    return dfs_memo(start)


# Note: No paths exist from dac to fft
svr_fft = count_paths_dag(DATA, "svr", "fft")
fft_dac = count_paths_dag(DATA, "fft", "dac")
dac_out = count_paths_dag(DATA, "dac", "out")

svr_fft * fft_dac * dac_out
```
