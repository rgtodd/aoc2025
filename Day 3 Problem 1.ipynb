## Define get_max_joltage

Determine the maximum joltage a bank can produce

```python
def get_max_joltage(bank: str) -> str:
    # Track the maximum battery digit seen from the right
    max_right_battery = ""
    # Initialize maximum joltage to "00" (minimum possible two-digit value)
    max_joltage = "00"

    # Iterate through the bank from right to left
    for battery in bank[::-1]:
        # If we have a previous battery to the right
        if max_right_battery != "":
            # Form a two-digit joltage by combining current battery with max right battery
            current_joltage = battery + max_right_battery
            # Update max_joltage if this combination is greater (string comparison)
            if current_joltage > max_joltage:
                max_joltage = current_joltage

        # Update the maximum battery digit seen so far from the right
        if battery > max_right_battery:
            max_right_battery = battery

    return max_joltage
```

## Compute Total Joltage

Total up the maximum joltages for each bank.

```python
# Initialize the total joltage counter
total_joltage = 0

# Process each battery bank and sum up their maximum joltages
for bank in DATA:
    total_joltage += int(get_max_joltage(bank))

total_joltage
```
