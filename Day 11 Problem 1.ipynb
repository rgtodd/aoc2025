## Load Source Data

Load source data into `DATA`.

```python
# Read the input file containing turn instructions
with open("data/day11.txt") as f:
    DATA = {values[0]: tuple(values[1:]) for line in f if (values := line.strip().replace(':', '').split())}

# DATA
```

## Create Graph and Find Minimum Distance

```python
import networkx as nx

G = nx.DiGraph()
for from_node, to_nodes in DATA.items():
    for to_node in to_nodes:
        G.add_edge(from_node, to_node)

source = "you"
target = "out"

paths = list(nx.all_simple_paths(G, source=source, target=target))
path_count = len(paths)

path_count

```
