## Load Source Data

Load source data into `DATA`.

```python
# Read the input file containing comma-separated ranges
with open("data/day2.txt") as f:
    DATA = f.read().split(',')

#DATA
```

## Parse Source Data

Parse each value into integer ranges.

```python
def parse_source_range(s: str) -> tuple[int, int]:
    """Parse a range string like '123-456' into a tuple of integers (123, 456)"""
    start, end = s.split('-')
    return (int(start), int(end))

# Convert all range strings to tuples of integers
RANGES = [parse_source_range(r) for r in DATA]

#RANGES
```

## Define is_valid Function

```python
def is_valid(id: int) -> bool:
    """
    Check if an ID is valid.
    An ID is INVALID if it consists of a repeating pattern.
    Examples: 
      - 111 is invalid (pattern '1' repeated 3 times)
      - 123123 is invalid (pattern '123' repeated 2 times)
      - 1234 is valid (no repeating pattern)
    """
    # Convert id to string
    id_str = str(id)
    id_str_len = len(id_str)

    # Determine the maximum substring length (can't be more than half the total length)
    max_substring_len = id_str_len // 2

    # Try each possible substring length
    for substring_len in range(1, max_substring_len + 1):
        # Only check if the substring divides evenly into the full string
        if id_str_len % substring_len == 0:
            repeat_count = id_str_len // substring_len
            # Check if repeating the substring recreates the original string
            if id_str == (id_str[:substring_len] * repeat_count):
                return False

    return True

is_valid(111)
```

## Count total of invalid ID's

```python
invalid_total = 0

# Iterate through all ranges and check each ID
for (lower_bound, upper_bound) in RANGES:
    for id in range(lower_bound, upper_bound + 1):
        if not is_valid(id):
            # Add invalid IDs to the total
            # print(f"Adding {id}.")
            invalid_total += id

invalid_total
```
