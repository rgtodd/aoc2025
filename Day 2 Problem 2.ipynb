## Load Source Data

Load source data into `DATA`.

```python
f = open("data/day2.txt", "r")
DATA = f.read().split(',')
f.close()

#DATA
```

## Parse Source Data

Parse each value into integer ranges.

```python
def parse_source_range(s: str) -> tuple[int, int]:
    elements = s.split('-')
    range = (int(elements[0]), int(elements[1]))
    return range

RANGES = list(map(parse_source_range, DATA))

#RANGES
```

## Define is_valid Function

```python
def is_valid(id: int) -> bool:

    # Convert id to string.
    #
    id_str = str(id)
    id_str_len = len(id_str)

    # Determine the maximum substring length.
    # 
    max_substring_len = id_str_len // 2

    for substring_len in range(1, max_substring_len + 1):
        # Ensure substring divides evenly into string
        if id_str_len % substring_len == 0:
            repeat_count = id_str_len // substring_len
            if id_str == (id_str[:substring_len] * repeat_count):
                return False

    return True

is_valid(111)
```

## Count total of invalid ID's

```python
invalid_total = 0

for (lower_bound, upper_bound) in RANGES:
    for id in range(lower_bound, upper_bound + 1):
        if not is_valid(id):
            # print(f"Adding {id}.")
            invalid_total += id

invalid_total
```
