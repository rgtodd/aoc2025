## Load Source Data

Load source data into `DATA`.

```python
from shapely import Point

# Read and parse the input data as shapely Points.
with open("data/day9.txt") as f:
    DATA: list[Point] = [
        Point(int(elements[0]), int(elements[1]))
        for line in f
        if (elements := line.strip().split(","))
    ]

# DATA
```

## Search for largest rectangle

Searches for largest rectangle spanning two input points.

```python
# Find the largest rectangle that can be formed by any two points
max_area = 0
max_bounds = None

# Compare each unique pair of points
for i, p1 in enumerate(DATA):
    for p2 in DATA[i + 1:]:
        # Determine rectangle corners from the two points
        x_min, x_max = min(p1.x, p2.x), max(p1.x, p2.x)
        y_min, y_max = min(p1.y, p2.y), max(p1.y, p2.y)

        # Calculate area (inclusive of boundary points)
        area = (int(x_max - x_min) + 1) * (int(y_max - y_min) + 1)

        # Track the largest rectangle found
        if area > max_area:
            max_area = area
            max_bounds = (x_min, y_min, x_max, y_max)

if max_bounds:
    print(f"max_area = {max_area}")
```
